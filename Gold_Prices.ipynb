```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('gold_monthly.csv')
df
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
#Checking for null values

pd.isnull(df).sum()
```

```python
#New DataFrames with monthly dates as index

date_range = pd.date_range(start = '1/1/1950' , end = '8/1/2020', freq = 'M')
df['month'] = date_range
df.drop('Date', axis = 1, inplace = True)
df = df.set_index('month')
df.head()
```

```python
#Plot gold prices over time
```

```python
plt.figure(figsize=(20,8))
df.plot(figsize = (20,8))
plt.title("Gold Price Monthly from 1950 and Onwards", fontweight='bold', fontsize=15)
plt.xlabel("Month" , fontweight='bold', fontsize=12)
plt.ylabel("Price" , fontweight='bold', fontsize=12)
plt.show()
```

```python
#Create a boxplot of the gold prices by year

plt.figure(figsize = (25, 10))

df['year'] = df.index.year

sns.boxplot(x= df.year, y=df.values[:,0] )

plt.title('Gold Price Yearly from 1950 and onwards', fontweight='bold', fontsize=18)
plt.xlabel('Year', fontweight='bold', fontsize=15)
plt.ylabel('Price', fontweight='bold', fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.show()

```

```python
#Summary statistics of the gold prices

print("Summary statistics of the gold prices:\n", df.describe())
```

```python
df_yearly_sum = df.resample('A').mean()
df_yearly_sum.plot()
plt.title('Avg Gold Price per Year since 1950')
plt.xlabel('Year')
plt.ylabel('Price')
plt.grid()
```

```python
df_quarterly_sum = df.resample('Q').mean()
df_quarterly_sum.plot()
plt.title('Avg Gold Price per Quarter since 1950')
plt.xlabel('Quarter')
plt.ylabel('Price')
plt.grid()
```
